```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados para os gráficos
dados = [
    {'To teach others to program': 21, 'To program in an uncomplicated way': 11, 'For personal objectives': 4},
    {'To teach others to program': 15, 'To program in an uncomplicated way': 8, 'For personal objectives': 6},
    {'To teach others to program': 25, 'To program in an uncomplicated way': 9, 'For personal objectives': 8},
    {'To teach others to program': 18, 'To program in an uncomplicated way': 14, 'For personal objectives': 10},
    {'To teach others to program': 20, 'To program in an uncomplicated way': 12, 'For personal objectives': 5},
    {'To teach others to program': 22, 'To program in an uncomplicated way': 10, 'For personal objectives': 7}
]

# Criando a figura e os subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Cores para as barras
cores = ['blue', 'green', 'red']

# Iterando sobre os dados e os subplots
for i, dado in enumerate(dados):
    # Calculando as posições do subplot na grade
    row = i // 3
    col = i % 3
    
    # Plotando o gráfico de barras
    barras = axs[row, col].bar(dado.keys(), dado.values(), color=cores)
    axs[row, col].set_xlabel('Answers')
    axs[row, col].set_ylabel('Quantity')
    axs[row, col].set_title(f'Graph {i+1}')
    axs[row, col].set_yticks(range(0, max(dado.values()) + 1, 5))
    axs[row, col].tick_params(axis='x', rotation=45)
    axs[row, col].legend(barras, dado.keys())

plt.tight_layout()
plt.show()
```
